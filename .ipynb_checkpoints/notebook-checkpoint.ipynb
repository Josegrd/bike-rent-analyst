## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana penjualan dalam 2 tahun terakhir?

```python
# rincian penyewaan keseluruhan 2011-2012
day_df.groupby(by="mnth").agg({
    "cnt": ["max", "min", "mean", "std"]
})
```

```python
# sum_order_df = day_df.groupby("mnth").cnt.sum().reset_index()
# sum_order_df.head(12)
```

```python
# rincian spesifik penyewaan tiap bulan pada 2011
filtered_day11_df = day_df[day_df['yr'] == 0]
sum_order11_df = filtered_day11_df.groupby(["yr", "mnth"]).cnt.sum().reset_index()
sum_order11_df.head(12)
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(sum_order11_df["mnth"], 
         sum_order11_df["cnt"], 
         marker='o', 
         linewidth=2, 
         color="#72BCD4") 
plt.title("Penyewaan Sepeda pada Tahun 2011", loc="center", fontsize=20) 
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```

statistik penyewaan sepeda pada Tahun 2011, terbilang cukup fluktuatif diawal, seperti terlihat pada grafik diatas, jumlah menyewaan meningkat dari bulan JANUARI hingga puncak nya di bulan JUNI (titik penyewaan tertinggi), setelah itu juli-desember, jumlah penjualan mulai menurun.
Dapat disimpulkan penyewaan terbanyak terjadi pada periode MEI-AGUSTUS 2011

```python
# rincian spesifik penyewaan tiap bulan pada 2012
filtered_day12_df = day_df[day_df['yr'] == 1]
sum_order12_df = filtered_day12_df.groupby(["yr", "mnth"]).cnt.sum().reset_index()
sum_order12_df.head(12)
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(sum_order12_df["mnth"], 
         sum_order12_df["cnt"], 
         marker='o', 
         linewidth=2, 
         color="#72BCD4") 
plt.title("Penyewaan Sepeda pada Tahun 2012", loc="center", fontsize=20) 
plt.xticks(fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```

Sementara itu, sama seperti 2011, statistik penyewaan sepeda pada Tahun 2012, terbilang cukup fluktuatif diawal, seperti terlihat pada grafik diatas, jumlah menyewaan meningkat dari bulan JANUARI hingga puncak nya di bulan SEPTEMBER (titik penyewaan tertinggi), setelah itu september-desember, jumlah penjualan mulai menurun.
Dapat disimpulkan penyewaan terbanyak terjadi pada periode JUNI-OKTOBER 2012

### Pertanyaan 2: Dibulan apa orang banyak menyewa sepeda?

```python
# kalkulasi rata-rata 2011-2012
sum_rent_month_df = day_df.groupby("mnth").cnt.sum().sort_values(ascending=False).reset_index()
sum_rent_month_df.head(12)
```

```python
colors = ["#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3","#D3D3D3", "#D3D3D3","#D3D3D3", "#72BCD4", "#D3D3D3", "#D3D3D3","#D3D3D3", "#D3D3D3", "#D3D3D3"]

 
plt.figure(figsize=(10, 5))
 
sns.barplot(
    y="cnt", 
    x="mnth",
    data= sum_rent_month_df.sort_values(by="cnt", ascending=False),
    palette=colors
)
plt.title("Rata-rata Penyewaan Setiap Bulan pada Tahun 2011-2012", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

kalkulasi penyewaan setiap bulannya pada 2011-2012, periode penyewaan sepeda terbanyak terjadi pada bulan JUNI-SEPTEMBER, dengan penyewaan terbanyak pada bulan AGUSTUS

```python
# kalkulasi rata-rata pada 2011
filtered_day11_df = day_df[day_df['yr'] == 0]
sum_order11_df = filtered_day11_df.groupby(["yr", "mnth"]).cnt.sum().reset_index()
sum_order11_df.head(12)
```

```python
colors = ["#D3D3D3"] * len(sum_order11_df) 
max_value_row11 = sum_order11_df[sum_order11_df['cnt'] == sum_order11_df['cnt'].max()]
max_index11 = max_value_row11.index[0]
colors[max_index11] = "#72BCD4" 
 
plt.figure(figsize=(10, 5))
 
sns.barplot(
    y="cnt", 
    x="mnth",
    data= sum_order11_df,
    palette=colors
)
plt.title("Jumlah Penyewaan Sepeda Pada Tahun 2012", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

Kalkulasi penyewaan sepeda terbanyak pada tahun 2011, terjadi pada bulan JUNI

```python
# kalkulasi rata-rata pada 2012
filtered_day12_df = day_df[day_df['yr'] == 1]
sum_order12_df = filtered_day12_df.groupby(["yr", "mnth"]).cnt.sum().reset_index()
sum_order12_df.head(12)
```

```python
colors = ["#D3D3D3"] * len(sum_order12_df) 
max_value_row12 = sum_order12_df[sum_order12_df['cnt'] == sum_order12_df['cnt'].max()]
max_index12 = max_value_row12.index[0]
colors[max_index12] = "#72BCD4" 
 
plt.figure(figsize=(10, 5))
 
sns.barplot(
    y="cnt", 
    x="mnth",
    data= sum_order12_df,
    palette=colors
)
plt.title("Jumlah Penyewaan Sepeda Pada Tahun 2012", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

Kalkulasi penyewaan sepeda terbanyak pada tahun 2012, terjadi pada bulan SEPTEMBER

### Pertanyaan 3: Dijam berapa orang paling ramai menyewa sepeda?

```python
filtered_hour_df =hour_df.groupby("hr").cnt.sum().reset_index()
filtered_hour_df.head(12)
```

```python
colors = ["#D3D3D3"] * len(filtered_hour_df) 
max_value_row = filtered_hour_df[filtered_hour_df['cnt'] == filtered_hour_df['cnt'].max()]
max_index = max_value_row.index[0]
colors[max_index] = "#72BCD4" 

plt.figure(figsize=(10, 5))
 
sns.barplot(
    y="cnt", 
    x="hr",
    data= filtered_hour_df,
    palette=colors
)
plt.title("Jumlah Penyewaan Sepeda Pada Tahun 2012", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

hasil kalkulasi dalam 2 Tahun terakhir, penyewaan terbanyak terjadi pada jam : 17 atau 5 Sore

### Pertanyaan 4: Musim apa paling ramai orang menyewa sepeda?

```python
filtered_season_df = day_df.groupby(by="season").cnt.sum().reset_index()
filtered_season_df.head()
```

```python
colors = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]

plt.figure(figsize=(10, 5))
 
sns.barplot(
    y="season", 
    x="cnt",
    data= filtered_season_df,
    palette=colors,
    orient="h"
)
plt.title("Jumlah Penyewaan Sepeda Menurut Cuaca", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)

plt.show()

```
